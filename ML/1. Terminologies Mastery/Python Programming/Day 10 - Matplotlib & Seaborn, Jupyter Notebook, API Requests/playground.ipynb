## Task 1 — run this cell first
This loads `df` into memory. Run it, confirm the output table appears INLINE, right below this cell — not in a separate window, not in your terminal.

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Varun", "Priya", "Raj"],
    "age": [30, 25, 45],
    "salary": [70000, 62000, 85000],
})

df.head()
```

## Task 2 — run this cell SEPARATELY, without re-running Task 1
This proves the "shared kernel" point from Theory.md ③ — `df` is already in memory from Task 1's cell, so this cell can reference it directly, with no re-import, no reload.

```python
df["age"].mean()
```

## Task 3 — a chart, rendered INLINE (no separate popup window this time)
Unlike running a `.py` script from the terminal (where `plt.show()` opens a native OS window), inside a notebook the chart renders directly below the cell, as part of the notebook itself.

```python
import matplotlib.pyplot as plt

plt.scatter(df["age"], df["salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
plt.show()
```

## Task 4 — the out-of-order trap (Theory.md ③'s real warning)
Run this cell now — it OVERWRITES `df` with only 2 rows. Then scroll UP and re-run Task 2's cell (`df["age"].mean()`) WITHOUT re-running Task 1.

Predict first: will Task 2 now use the original 3-row `df`, or this new 2-row one? Then check the execution-order number in brackets (e.g. `[4]`) next to each cell to confirm WHY.
